Importing the libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn import svm
```

Data collection annd analysis
PIMA diabetes dataset

```python
dataset=pd.read_csv('diabetes.csv')
dataset.head()
```

number of rows and columns in dataset

```python
dataset.shape
```

Mathematical measures of dataset

```python
dataset.describe()
```

```python
# get count of different outcomes
dataset['Outcome'].value_counts()
```

```python
dataset.groupby('Outcome').mean()
```

```python
# Seperate the label
X=dataset.drop(columns='Outcome',axis=1)
Y=dataset['Outcome']
print(X.shape)
print(Y.shape)
```

Data Standardization

```python
scaler = StandardScaler()

#feed data to scaler instance of above class
scaler.fit(X)

standardized_data=scaler.transform(X)
print(standardized_data)
# now data is in the range of 0-1
```

```python
#we can reassign X with modified data
X = standardized_data
Y = dataset['Outcome']
```

Splitting data for testing and training

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, Y_train.shape, X_test.shape)
```

Training Model

```python
classifier = svm.SVC(kernel='linear')
```

```python
# training the support vector machine classifier
classifier.fit(X_train, Y_train)
```

```python
# checking accuracy score on training data
prediction_on_training_dataset = classifier.predict(X_train)
accuracy_score_on_training_dataset = accuracy_score(prediction_on_training_dataset, Y_train)
print('Accuracy on training dataset : ',accuracy_score_on_training_dataset*100,"%")
```

```python
# Checking paccuracy on testing dataset
prediction_on_test_dataset = classifier.predict(X_test)
accuracy_on_test_dataset = accuracy_score(prediction_on_test_dataset, Y_test)
print('Accuracy on test dataset : ',accuracy_on_test_dataset*100,'%')
```

```python
import gradio as gr

def check_diabetes(input):
  input_as_list=input.split(',')
  input_as_float=[float(item) for item in input_as_list]
  input_as_array=np.asarray(input_as_float)
  input_reshaped=input_as_array.reshape(1,-1)

  #standardizing data
  input_satandardized = scaler.transform(input_reshaped)
  prediction=classifier.predict(input_satandardized)

  if(prediction=='0'):
    return "The person is not diabetic"
  else:
    return "The person is diabetic"

ui = gr.Interface(fn=check_diabetes , inputs='text', outputs='text')
ui.launch()
```
